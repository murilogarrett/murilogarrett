# Pixel e Estrutua de dados de imagens

**Pixel** é a menor unidade de informação que compoe uma imagem digital. 
Em uma imagem **RGB** o pixel é representado por tres numeros de **8 bits** associados com as cores **R, G e B **que correspondem as cores **Red, Green e Blue** respectivamente.
Como são numeros de 8 bits a faixa de valores é de 0 até 255.

```python
Image.open("/content/sample_data/imagens/pixel-RGB.jpg")
```

A cor do pixel será gerada a partir da combinação de cada um dos canais. Se a combinação de valores for 255 a cor do pixel será **branco**, se a combinação for 0, a cor do Piexel será **preto**.

# Tecnicas de pré-processamento aplicado a imagens

Normalizando valores de pixels

```python
image = Image.open("/content/sample_data/imagens/boi.jpeg")
```

convertndo a imagem em um array


```python
pixels = asarray(image)
```

verificando valores entre 0 e 255

```python
print('Data Type: %s' % pixels.dtype)
print('Min: %.3f, Max: %.3f' % (pixels.min(), pixels.max()))
```

converte os valores inteiros em float pra realizar a operação

```python
pixels = pixels.astype('float32')
```

normaliza a faixa de valores

```python
pixels /= pixels.max()
```

Verificando a faixa de valores normalizada

```python
print('Data Type: %s' % pixels.dtype)
print('Min: %.3f, Max: %.3f' % (pixels.min(), pixels.max()))
```
